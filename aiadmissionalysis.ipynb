```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load the data from the CSV file
file_path = r"D:\Brave Downloads\student_names_with_year_gender.csv"
data = pd.read_csv(file_path)

# Check if 'Student Name' column exists
if 'Student Name' not in data.columns:
    print("The 'Student Name' column is missing from the data.")
else:
    # Extract the first letter of each name
    data['First Letter'] = data['Student Name'].str[0].str.upper()  # Convert to uppercase for uniformity

    # Calculate frequency of first letters
    first_letter_counts = data['First Letter'].value_counts(normalize=True).sort_index() * 100  # Convert to percentage

    # English letter frequencies (in percentage)
    english_letter_frequencies = {
        'A': 8.2,
        'B': 1.5,
        'C': 2.8,
        'D': 4.3,
        'E': 12.7,
        'F': 2.2,
        'G': 2.0,
        'H': 6.1,
        'I': 7.0,
        'J': 0.15,
        'K': 0.8,
        'L': 4.0,
        'M': 2.4,
        'N': 6.7,
        'O': 7.5,
        'P': 1.9,
        'Q': 0.1,
        'R': 6.0,
        'S': 6.3,
        'T': 9.1,
        'U': 2.8,
        'V': 1.0,
        'W': 2.4,
        'X': 0.2,
        'Y': 2.0,
        'Z': 0.1
    }

    # Create a DataFrame for comparison
    comparison_df = pd.DataFrame({
        'Student Name Frequency (%)': first_letter_counts,
        'English Letter Frequency (%)': pd.Series(english_letter_frequencies)
    }).fillna(0)  # Fill NaN values with 0 for letters not present in student names

    # Calculate Cosine Similarity
    cosine_sim = cosine_similarity([comparison_df['Student Name Frequency (%)'].values], 
                                   [comparison_df['English Letter Frequency (%)'].values])[0][0]

    # Calculate Euclidean Distance
    euclidean_distance = np.linalg.norm(comparison_df['Student Name Frequency (%)'].values - 
                                        comparison_df['English Letter Frequency (%)'].values)

    # Display the results
    print(f"Cosine Similarity: {cosine_sim:.4f}")
    print(f"Euclidean Distance: {euclidean_distance:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample student admission data
admissions_data = {
    'Year': [2019, 2020, 2021, 2022, 2023, 2024],
    'Admissions': [15, 8, 8, 5, 12, 26]
}

# Create a DataFrame for admissions
admissions_df = pd.DataFrame(admissions_data)

# Load COVID-19 data for India
covid_data_path = "D:\FDM downloads\WHO-COVID-19-global-daily-data.csv"
covid_df = pd.read_csv(covid_data_path)

# Filter COVID-19 data for India
india_covid_df = covid_df[covid_df['Country'] == 'India']

# Convert 'Date_reported' to datetime
india_covid_df['Date_reported'] = pd.to_datetime(india_covid_df['Date_reported'])

# Extract year and sum new cases by year
india_covid_df['Year'] = india_covid_df['Date_reported'].dt.year
covid_yearly_data = india_covid_df.groupby('Year')['New_cases'].sum().reset_index()

# Merge admissions data with COVID-19 data
merged_data = pd.merge(admissions_df, covid_yearly_data, on='Year', how='outer')
merged_data.columns = ['Year', 'Admissions', 'New_Cases']

# Plotting
plt.figure(figsize=(12, 6))

# Plot admissions
plt.subplot(1, 2, 1)
plt.bar(merged_data['Year'], merged_data['Admissions'], color='skyblue', label='Admissions')
plt.title('Student Admissions (2019-2024)')
plt.xlabel('Year')
plt.ylabel('Number of Admissions')
plt.xticks(merged_data['Year'])
plt.grid(axis='y')

# Plot COVID-19 new cases
plt.subplot(1, 2, 2)
plt.plot(merged_data['Year'], merged_data['New_Cases'], marker='o', color='orange', label='New COVID-19 Cases')
plt.title('New COVID-19 Cases in India (Yearly)')
plt.xlabel('Year')
plt.ylabel('Number of New Cases')
plt.xticks(merged_data['Year'])
plt.grid()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample student admission data
admissions_data = {
    'Year': [2019, 2020, 2021, 2022, 2023, 2024],
    'Admissions': [15, 8, 8, 5, 12, 26]
}

# Create a DataFrame for admissions
admissions_df = pd.DataFrame(admissions_data)

# Load COVID-19 data for India
covid_data_path = "D:\FDM downloads\WHO-COVID-19-global-daily-data.csv"
covid_df = pd.read_csv(covid_data_path)

# Filter COVID-19 data for India
india_covid_df = covid_df[covid_df['Country'] == 'India']

# Convert 'Date_reported' to datetime
india_covid_df['Date_reported'] = pd.to_datetime(india_covid_df['Date_reported'])

# Extract year and sum new cases by year
india_covid_df['Year'] = india_covid_df['Date_reported'].dt.year
covid_yearly_data = india_covid_df.groupby('Year')['New_cases'].sum().reset_index()

# Merge admissions data with COVID-19 data
merged_data = pd.merge(admissions_df, covid_yearly_data, on='Year', how='outer')
merged_data.columns = ['Year', 'Admissions', 'New_Cases']

# Calculate correlation
correlation = merged_data['Admissions'].corr(merged_data['New_Cases'])
print(f"Correlation between Admissions and New COVID-19 Cases: {correlation:.4f}")

# Plotting
plt.figure(figsize=(12, 6))

# Scatter plot for correlation
sns.scatterplot(data=merged_data, x='New_Cases', y='Admissions', color='blue', s=100)
plt.title('Correlation between Student Admissions and New COVID-19 Cases in India')
plt.xlabel('New COVID-19 Cases')
plt.ylabel('Student Admissions')
plt.grid()

# Add a regression line
sns.regplot(data=merged_data, x='New_Cases', y='Admissions', scatter=False, color='orange')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample student admission data
admissions_data = {
    'Year': [2019, 2020, 2021, 2022, 2023, 2024],
    'Admissions': [15, 8, 8, 5, 12, 26]
}

# Create a DataFrame for admissions
admissions_df = pd.DataFrame(admissions_data)

# Load AI development data
ai_data_path = 'notable_ai_models.csv'
ai_df = pd.read_csv(ai_data_path)

# Convert 'Publication date' to datetime and extract year
ai_df['Publication date'] = pd.to_datetime(ai_df['Publication date'], errors='coerce')
ai_df['Year'] = ai_df['Publication date'].dt.year

# Aggregate AI development data by year
ai_yearly_data = ai_df.groupby('Year').size().reset_index(name='AI_Developments')

# Merge admissions data with AI development data
merged_data = pd.merge(admissions_df, ai_yearly_data, on='Year', how='outer')

# Calculate correlation
correlation = merged_data['Admissions'].corr(merged_data['AI_Developments'])
print(f"Correlation between Admissions and AI Developments: {correlation:.4f}")

# Plotting
plt.figure(figsize=(12, 6))

# Scatter plot for correlation
sns.scatterplot(data=merged_data, x='AI_Developments', y='Admissions', color='blue', s=100)
plt.title('Correlation between Student Admissions and AI Developments in India')
plt.xlabel('Number of AI Developments')
plt.ylabel('Student Admissions')
plt.grid()

# Add a regression line
sns.regplot(data=merged_data, x='AI_Developments', y='Admissions', scatter=False, color='orange')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample student admission data
admissions_data = {
    'Year': [2019, 2020, 2021, 2022, 2023, 2024],
    'Admissions': [15, 8, 8, 5, 12, 26]  # Number of students
}

# Create a DataFrame for admissions
admissions_df = pd.DataFrame(admissions_data)

# Sample salary data (in INR)
salary_data = {
    'Year': [2019, 2020, 2021, 2022, 2023, 2024],
    'Salary': [600000, 700000, 800000, 900000, 1000000, 1200000]  # Average salary in INR
}

# Create a DataFrame for salaries
salary_df = pd.DataFrame(salary_data)

# Merge admissions data with salary data
merged_data = pd.merge(admissions_df, salary_df, on='Year', how='outer')

# Calculate correlation
correlation = merged_data['Admissions'].corr(merged_data['Salary'])
print(f"Correlation between Admissions and Salaries: {correlation:.4f}")

# Plotting
plt.figure(figsize=(10, 6))

# Scatter plot
sns.scatterplot(data=merged_data, x='Salary', y='Admissions', color='blue', s=100)
plt.title('Scatter Plot of Student Admissions vs. Average Salaries (2019-2024)')
plt.xlabel('Average Salary (INR)')
plt.ylabel('Number of Admissions')
plt.grid()

# Add a regression line
sns.regplot(data=merged_data, x='Salary', y='Admissions', scatter=False, color='orange')

# Show the plot
plt.tight_layout()
plt.show()
```
